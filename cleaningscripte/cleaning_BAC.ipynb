```python
# Importing useful libraries
import pandas as pd
import seaborn as sns
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv("./.env"))
path2 = os.environ.get("IPSLYC")
path3 = os.environ.get("EXPDIR")
# read datasets
df_bac1 = pd.read_csv(os.path.join(path2, "fr-en-indicateurs-de-resultat-des-lycees-denseignement-general-et-technologique.csv"), sep=";")
df_bac2 = pd.read_csv(os.path.join(path2, "fr-en-indicateurs-de-resultat-des-lycees-denseignement-professionnels.csv"), sep=";")
```

```python
df_bac1.head()
```

```python
df_bac2.head()
```

```python
df_bac1["UAI"].nunique()
```

```python
# Ne gardons que les résultats supérieur à 2016 avec les années d'IPS
# EGT et EP
df_bac1 = df_bac1[df_bac1.Annee > 2016]
df_bac2 = df_bac2[df_bac2.annee > 2016]
```

```python
# Ne gardons que les colonnes utiles avec le fichier IPS
# EGT
df_bac1 = df_bac1[['Annee', 'UAI', 'Etablissement', 'Ville', 'Code commune', 'Academie', 'Departement', 'Secteur', 'Presents - Toutes series','Taux de reussite - Toutes series']]
```

```python
# Ne gardons que les colonnes utiles avec le fichier IPS
# EP
df_bac2 = df_bac2[['annee', 'code_etablissement', 'etablissement', 'ville', 'commune', 'academie', 'departement', 'secteur_public_1_prive_2', 'effectif_presents_total_secteurs', 'taux_brut_de_reussite_total_secteurs']]
```

```python
df_bac1.info()
```

```python
df_bac2.info()
```

```python
colNames = df_bac2.columns.to_list()
```

```python
df_bac1.columns = colNames
```

```python
df_bac1.info()
```

```python
df_bac2.info()
```

```python
# Concaténation
df_bac = pd.concat([df_bac1 ,df_bac2], axis=0)
df_bac = df_bac.reset_index()
```

```python
df_bac.head()
```

```python
# creation INDEX unique Etablissement / Année
df_bac['UniqueIndex'] = df_bac['code_etablissement'] + df_bac['annee'].astype('str')
```

```python
# export data
#path3 = os.environ.get("EXPDIR")
df_bac.to_csv(os.path.join(path3,"resultat_bac_2017_2023.csv"))
```
