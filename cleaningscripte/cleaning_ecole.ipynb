```python
import pandas as pd
import seaborn as sns
import os
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv("./.env"))
path1 = os.environ.get("IPSECOL")
path_ap22 = os.environ.get("IPSECOL_AP2022")
path2 = os.environ.get("CLOSED")
path3 = os.environ.get("GEOCOORD")
path4 = os.environ.get("DESTFILE_1622")
```

```python
# reading the datasets
df_2016 = pd.read_csv(path1, sep=";")
df_2022 = pd.read_csv(path_ap22, sep=";")
```

```python
# Renomer la colonne "nom_de_l_etablissment" en "nom_de_l_etablissement" pour que ce soit le meme titre aue sur df_2022
df_2016.rename(columns={"nom_de_l_etablissment": "nom_de_l_etablissement"}, inplace=True)
```

```python
# Formater la date dans les deux datasets
for df in [df_2016, df_2022]:
    df["rentree_scolaire"] = df["rentree_scolaire"].str.split("-").str[0]
```

```python
df_2022.head()
```

```python
df_2022.info()
```

```python
df_2016.head()
```

```python
# Fusionner les deux datasets
df = pd.concat([df_2016, df_2022], ignore_index=True)
```

```python
# Sauvegarder le dataset fusionné
selected_columns=["rentree_scolaire", "uai", "nom_de_l_etablissement", "secteur", "ips", "effectifs"]
df_ecoles = df[selected_columns]
df_ecoles.to_csv(path4 + "\ecoles_1622.csv", index=False)
print("DataFrame fusionné sauvegarde sur:", path4 + "\ecoles_1622.csv")
```
