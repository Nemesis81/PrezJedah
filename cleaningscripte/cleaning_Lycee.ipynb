Normalisation des data 2016-2022 
- Normalisé_x = (x - min(x)) / (max(x) - min(x))

```python
df_2016['ips_ensemble_gt_pro']= df_2016['ips_ensemble_gt_pro'] / df_2016['ips_ensemble_gt_pro'].abs().max()
df_2022['ips_ensemble_gt_pro']= df_2022['ips_ensemble_gt_pro'] / df_2022['ips_ensemble_gt_pro'].abs().max()
df_2023['ips_ensemble_gt_pro']= df_2023['ips_ensemble_gt_pro'] / df_2023['ips_ensemble_gt_pro'].abs().max()
```

```python
# split colonne de rentrée scolaire en colonne de rentrée et colonne de diplôme.
# Changement du type en integer au lieu de string.
df_1623[['annee_rentree','annee_diplome']]=df_1623['rentree_scolaire'].str.split("-", expand=True)
df_1623['annee_rentree'] = df_1623['annee_rentree'].astype('int')
df_1623['annee_diplome'] = df_1623['annee_diplome'].astype('int')
```

```python
# creation INDEX unique Etablissement / Année
df_1623['UniqueIndex'] = df_1623['uai'] + df_1623['annee_diplome'].astype('str')
```
